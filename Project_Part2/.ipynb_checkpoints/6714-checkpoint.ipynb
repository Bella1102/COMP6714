```python
import spacy
import math
from collections import defaultdict
class InvertedIndex:
    
    def __init__(self):
        ## You should use these variable to store the term frequencies for tokens and entities...
        self.tf_tokens=defaultdict(dict,{})    #save the token index
        self.tf_entities=defaultdict(dict,{})  #save the entity index

        ## You should use these variable to store the inverse document frequencies for tokens and entities...
        self.idf_tokens=defaultdict(dict,{}) 
        self.idf_entities=defaultdict(dict,{}) 

    ## Your implementation for indexing the documents...
    def index_documents(self, documents):
        ## Replace this line with your implementation...
        self.documents=documents

        spy = spacy.load("en_core_web_sm")
        
        spy_doc={}
        remtoken={}
        for id,value in documents.items():    # use text (string variable)  saving the text of document
            doc = spy(value.replace('\n',' '))
            spy_doc[id]=doc
            entit=[]
            toke=[]
            for ent in doc.ents:
                entit.append(ent.text)
            for ele in set(entit):
                if ele not in self.tf_entities.keys():
                    tmp={}
                    tmp[id]=entit.count(ele)
                    self.tf_entities[ele]=tmp
                else:
                    self.tf_entities[ele][id]=entit.count(ele) 
            for tok in doc:
                if tok.is_stop==False and tok.is_punct==False: 
                    toke.append(tok.text)
            
            for ele in set(entit):
                if ele in toke:
                    if ele not in remtoken.keys():
                        tmp={}
                        tmp[id]=entit.count(ele)
                        remtoken[ele]=tmp
                    else:
                        remtoken[ele][id]=entit.count(ele)
                    
        
        for id,value in documents.items():
            doc = spy_doc[id]
            toke=[]
            for tok in doc:
                if tok.is_stop==False and tok.is_punct==False: 
                    toke.append(tok.text)
            for tok in set(toke):
                if tok not in self.tf_tokens.keys():
                    tmp={}
                    tmp[id]=toke.count(tok)
                    self.tf_tokens[tok]=tmp
                else:
                    self.tf_tokens[tok][id]=toke.count(tok)
        for key,value in remtoken.items():
            for id,count in value.items():
                self.tf_tokens[key][id]-=remtoken[key][id]
                if self.tf_tokens[key][id]==0:
                    self.tf_tokens[key].pop(id)
            tmp={}
            if self.tf_tokens[key]==tmp:
                self.tf_tokens.pop(key)
            
                        
        for ent in self.tf_entities.keys():
            cnt=len(self.tf_entities[ent])
            self.idf_entities[ent]=1+math.log(len(self.documents)/(cnt+1))
                       
        for tok in self.tf_tokens.keys():
            cnt=len(self.tf_tokens[tok])
            self.idf_tokens[tok]=1+math.log(len(self.documents)/(cnt+1))
            
```

```python
traindata=[]
train_group=[]
for key,value in train_mentions.items():
    train_group.append(len(value['candidate_entities']))
    for ents in value['candidate_entities']:
        entsplit=ents.split('_')
        length=len(entsplit)
        s1=0
        s2=0
        s3=0
        for ent in entsplit:           
            if train_mentions[key]['doc_title'] in index.tf_tokens[ent].keys():
                tf_tok=1+math.log(index.tf_tokens[ent][train_mentions[key]['doc_title']])
                idf_tok=index.idf_tokens[ent]
                s1+=tf_tok*idf_tok
        ent=ents.replace('_',' ')   
        if train_mentions[key]['doc_title'] in index.tf_entities[ent].keys():
            tf_ent=1+math.log(1+math.log(index.tf_entities[ent][train_mentions[key]['doc_title']]))
            idf_ent=index.idf_entities[ent]
            s2+=tf_ent*idf_ent
        common_len=len([term for term in ents.lower().split('_') if term in value['mention'].lower().split(' ')])
        rate=common_len/length
        
        ent2=value['mention'].split()
        for ent in ent2:           
            if train_mentions[key]['doc_title'] in index.tf_tokens[ent].keys():
                tf_tok=1+math.log(index.tf_tokens[ent][train_mentions[key]['doc_title']])
                idf_tok=index.idf_tokens[ent]
                s3+=tf_tok*idf_tok
        length2=len(ent2)
        traindata.append([s1,s3,length,common_len,rate,length2])
```

```python
label=[]
for key,value in train_mentions.items():
    for cad in value['candidate_entities']:
        if cad==train_labels[key]['label']:
            label.append(1)
        else:
            label.append(0)
```

```python
import xgboost as xgb
import numpy as np
```

```python
def transform_data(features, groups, labels=None):
    xgb_data = xgb.DMatrix(data=features, label=labels)
    xgb_data.set_group(groups)
    return xgb_data
```

```python
traindata=np.array(traindata)
devdata=np.array(devdata)
xgboost_train = transform_data(traindata, train_group, label)
xgboost_test = transform_data(devdata, dev_group)
```

```python
param = {'max_depth': 8, 'eta': 0.05, 'silent': 1, 'objective': 'rank:pairwise',
         'min_child_weight': 0.01, 'lambda':100}
param['eval_metric'] = ['ndcg@5']
evallist = [ (xgboost_train, 'train') ]
classifier = xgb.train(param, xgboost_train,num_boost_round=4900,evals=evallist, early_stopping_rounds=10)
##  Predict test data...
preds = classifier.predict(xgboost_test)
preds
```
